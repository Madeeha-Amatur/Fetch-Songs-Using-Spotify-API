```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
client_id = '6f97bdf9bc43425fa20a4731f962dcdc'
client_secret = '34012d03e4a64e48b9d3619ee4b0df8a'
spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials(client_id, client_secret))
```

```python
#Function to convert duration of song in seconds to a duration timestamp MM:SS
def songlength(length_in_ms):
    seconds, length_in_ms = divmod(length_in_ms,1000)
    minutes, seconds = divmod(seconds, 60)
    length = f'{int(minutes):01d}:{int(seconds):02d}'
    return length
```

```python
#Fetch NewJeans albums from spotify
results = spotify.artist_albums('https://open.spotify.com/artist/6HvZYsbFfjnjFrWF950C9d')
albums = results['items']
while results['next']:
    results = spotify.next(results)
    albums.extend(results['items'])
    
all_songs = []
#get album details for each album
for album in albums:
    album_meta = spotify.album(album['id'])
    album = album_meta['name']
    artist = album_meta['artists'][0]['name']
    album_type = album_meta['album_type']
    release_date = album_meta['release_date']
    album_popularity = album_meta['popularity']
    results2 = album_meta['tracks']
    songs = results2['items']
    while results2['next']:
        results2 = spotify.next(results2)
        songs.extend(results2['items'])
    #get song details for each song within every album
    for song in songs:
        meta = spotify.track(song['id'])
        features = spotify.audio_features(song['id'])
        name = meta['name']
        trackNumber = meta['track_number']
        song_embed_preview_link = "https://embed.spotify.com/?uri=" + meta['uri']
        cover_image = meta['album']['images'][0]['url']
        length = songlength(meta['duration_ms'])
        song_popularity = meta['popularity']
        song_spotify_link = meta['external_urls']['spotify']
        acousticness = features[0]['acousticness']
        danceability = features[0]['danceability']
        energy = features[0]['energy']
        instrumentalness = features[0]['instrumentalness']
        liveness = features[0]['liveness']
        loudness = features[0]['loudness']
        speechiness = features[0]['speechiness']
        tempo = features[0]['tempo']
        valence = features[0]['valence']
        track = [artist, name, trackNumber, song_embed_preview_link, cover_image, album, album_type, album_popularity, release_date, length, song_popularity, song_spotify_link, acousticness, danceability, energy, 
                            instrumentalness, liveness, loudness, speechiness, tempo, valence]
        all_songs.append(track)
    
```

```python
#convert list to pandas dataframe
newJeans_songs = pd.DataFrame(all_songs,columns = ['artist', 'name', 'trackNumber', 'song_embed_preview_link', 'cover_image', 'album', 'album_type', 'album_popularity', 'release_date', 'length', 'song_popularity', 'song_spotify_link', 'acousticness', 'danceability', 'energy', 
                            'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence'] )
```

```python
#Sort NewJeans songs based on their popularity
newJeans_songs = newJeans_songs.sort_values('song_popularity', ascending=False, ignore_index=True)
```
